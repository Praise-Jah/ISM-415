```python
#import our pandas and matplotlib library

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#To read a data set

imdb = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
imdb.head()
```

```python
imdb.isnull().sum()
```

```python
imdb.dropna(how = 'any', inplace=True)
```

```python
imdb.isnull().sum()
```

# Answering Business Questions

### The Least Watched Movies

```python
#Groups movies tiltle and votes

imdb_votes = imdb.groupby('Title')['Votes'].sum().sort_values(ascending=False)
```

```python
#To show the graph of the least watched movies

plt.figure(figsize=(6,9))
imdb_votes[-20:].plot(kind='bar')
```

### The Most Watched Movies

```python
#To show the graph of the most watched movies

plt.figure(figsize=(5,7))
imdb_votes[:20][::-1].plot(kind='bar')
```

### The Least Rated Movies

```python
#Groups movies tiltle and rating

imdb_rating = imdb.groupby('Title')['Rating'].sum().sort_values(ascending=False)
```

```python
#To show the graph of the least rated movies

plt.figure(figsize=(7,9))
imdb_rating[-20:].plot(kind='bar')
```
