```python
import pandas as pd
```

```python
#To read a data set

data =pd.read_csv('NYC_Property_Data.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data
```

```python
#To drop missing values

data.dropna(how ='any', inplace = True)
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis=0, subset=['SALE PRICE', 'LAND SQUARE FEET', 'GROSS SQUARE FEET'])
```

```python
data = pd.get_dummies(data=data, columns=['BOROUGH','TAX CLASS AT TIME OF SALE','BUILDING CLASS AT TIME OF SALE'])
```

```python
A = data.drop('SALE PRICE', axis=1)
```

```python
B = data['SALE PRICE']
```

```python
B
```

```python
#To import the train test split

from sklearn.model_selection import train_test_split
```

```python
#Assign variables to split

A_train, A_test, B_train, B_test = train_test_split(A,B, test_size = 0.2, random_state = 101)
```

```python
A_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
model = regressor.fit(A_train, B_train)
```

```python
B_pred = model.predict(A_test)
```

```python
pd.DataFrame({'Actual':B_test, 'Predicted': B_pred, 'Error': B_test-B_pred})
```
